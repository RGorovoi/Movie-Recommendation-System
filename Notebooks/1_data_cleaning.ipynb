```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Merging the TMDB 5000 Movie Dataset (tmdb_5000_movies.csv) and TMDB 5000 Credits Dataset (tmdb_5000_credits.csv)
movies_df = pd.read_csv('../Data/tmdb_5000_movies.csv')
credits_df = pd.read_csv('../Data/tmdb_5000_credits.csv')

merged_df = pd.merge(movies_df, credits_df, left_on='id', right_on='movie_id', how='inner')
merged_df.drop('movie_id', axis=1, inplace=True)
```

```python
# Set an ID column
merged_df.set_index('id', inplace=True)
```

```python
merged_df.head(5)
```

```python
merged_df.info()
```

```python
# 3 values are missing for overview, 2 for runtime, and 1 for release date
# We can fill those with empty strings.
merged_df['tagline'] = merged_df['tagline'].fillna('')
merged_df['overview'] = merged_df['overview'].fillna('')
```

```python
# There are more than half of missing values of "homepage" column. 
# It's just links to website where you could buy/rent movies.
# We can fill the missing values with empty strings.
merged_df['homepage'] = merged_df['homepage'].fillna('')
```

```python
# Convert release dates to date-time format for easier filtering
merged_df['release_date'] = pd.to_datetime(merged_df['release_date'], errors='coerce')
merged_df = merged_df.dropna(subset=['release_date'])  # Drop one missing row
```

```python
# Also, there are some duplicate columns such as title_x and title_y, so remove one of them
merged_df = merged_df.drop('title_y', axis=1)
```

```python
# Rename column title_x to just title
merged_df = merged_df.rename(columns={'title_x': 'title'})
```

```python
# There is a column called "original_title" which is in the movie's original language
# I think it's reasonable to drop it and only leave the "title" column which is in English
merged_df = merged_df.drop('original_title', axis=1)
```

```python
# There are 2 values missing from runtime, which is the movie's length in minutes
# Replace it with the median value of the column
merged_df['runtime'] = merged_df['runtime'].fillna(merged_df['runtime'].median())
```
